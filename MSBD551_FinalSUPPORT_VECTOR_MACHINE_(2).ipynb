**SUPPORT VECTOR MACHINE**

```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Prepare the data
X = merged_df.drop(columns=['age','al','rbc','pc','pcc','ba','wbcc','rbcc','dm','cad','appet', 'pe','ane','class','htn','su','sc'])
y = merged_df['class']  # Target variable


#  Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size=0.2, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Hyperparameter tuning for SVM
svm_params = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale', 'auto'],
    'degree': [2, 3, 4]
}

svm_grid = GridSearchCV(SVC(random_state=101), svm_params, cv=5, scoring='accuracy')
svm_grid.fit(X_train, y_train)

# Evaluate best SVM model and display accuracy
best_svm = svm_grid.best_estimator_
y_pred_train = best_svm.predict(X_train)
y_pred_test = best_svm.predict(X_test)
train_acc = accuracy_score(y_true=y_train, y_pred=y_pred_train)
test_acc = accuracy_score(y_true=y_test, y_pred=y_pred_test)

print(f"SVM Model - Best Params: {best_svm.get_params()}")
print(f"SVM Model - Train Accuracy: {train_acc:.3f}, Test Accuracy: {test_acc:.3f}\n")

```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score, precision_score, recall_score, f1_score

# Assuming you have y_test (true labels) and y_pred_test (predicted labels)

# Print the classification report
report = classification_report(y_test, y_pred_test)
print("Classification Report:\n", report)

# Print the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_test)
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_test)

# Extract values from the confusion matrix
TN = conf_matrix[0, 0]  # True Negatives
FP = conf_matrix[0, 1]  # False Positives

# Calculate specificity
specificity = TN / (TN + FP)

print(f"Specificity: {specificity:.3f}")
```

```python
from sklearn.model_selection import cross_val_score


cv_scores = cross_val_score(best_svm, X, y, cv=5, scoring='accuracy')  # 5-fold cross-validation

print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean Accuracy: {cv_scores.mean():.3f}")
print(f"Standard Deviation: {cv_scores.std():.3f}")
```
